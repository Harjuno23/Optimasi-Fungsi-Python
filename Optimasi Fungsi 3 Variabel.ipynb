# Penerapan Genetic Algorithm pada Optimasi Fungsi 3 Variabel


```python
import numpy as np
import pandas as pd
```

```python
# batas atas dan batas bawah variabel
x_min = -10
x_max = 15
y_min = -7
y_max = 11
z_min = -20
z_max = 20

# jumlah kromosom pada populasi
n = 200

# generasi maksimal
N = 500
```

```python
# membentuk populasi 
def populasi():
    x = np.random.rand(n)*(x_max - x_min) + x_min
    y = np.random.rand(n)*(y_max - y_min) + y_min
    z = np.random.rand(n)*(z_max - z_min) + z_min
    pop = pd.DataFrame(
        {'x' : x,
         'y' : y,
         'z' : z
        }
    )
    
    return pop
```

```python
# nilai fitness fungsi
def fitness(pop):
    fitness = (pop['x'] + 2*pop['y'] - 10)**2 + (2*pop['x'] + pop['y'] + 5)**2 + 3*(pop['y']+pop['z'])
    pop['fitness'] = fitness
    
    return pop
```

```python
# seleksi secara acak
def seleksi():
    position = np.random.permutation(n)
    
    return position[0], position[1]
```

```python
# crossover rata2 2 kromosom
def crossover(pop):
    popc = pop.copy()
    for i in range(n):
        a, b = seleksi()
        x = (pop.loc[a] + pop.loc[b])/2
        popc.loc[i] = x
    
    return popc
```

```python
# mutasi dengan mengurangi setiap gen dengan bilangan acak antara -0.25 sampai 0.25 
def mutasi(popc, i):
    popm = popc.copy()

    for i in range(n):
        for j in popm.columns:
            popm.loc[i][j] = popm.loc[i][j] - (np.random.rand() - 0.5)/(i+1)

    return popm
```

```python
#mengabungkan populasi dengan hasil crossover dan mutasi
def gabung(pop, popc, popm):
    popAll = pop.copy()
    popAll = popAll.append(popc)
    popAll = popAll.append(popm)

    popAll = popAll.drop_duplicates()

    popAll.index = range(len(popAll))

    return popAll
```

```python
# mengurutkan berdasar nilai fitness
def urut(popAll):
    popAll = popAll.sort_values(by=['fitness'])
    
    popAll.index = range(len(popAll))

    return popAll
```

```python
#mengiliminasi populasi
def eliminasi(popAll):
    pop = popAll.head(n)
    
    return pop
```

```python
# jika nilai gen melebihi batasnya looping terhenti
def stop(pop):
    hasil = pop.loc[0][0]>x_max or pop.loc[0][1]>y_max or pop.loc[0][2]>z_max or pop.loc[0][0]<x_min or pop.loc[0][1]<y_min or pop.loc[0][2]<z_min
    return hasil
```

```python
# algoritma
pop = populasi()
pop = fitness(pop)
print('Generasi 0 (Initial)')
print(pop)

for i in range(1, N+1):
    popc = crossover(pop)
    popc = fitness(popc)

    popm = mutasi(popc, i)
    popm = fitness(popm)

    popAll = gabung(pop, popc, popm)

    popAll = urut(popAll)

    pop = eliminasi(popAll)
    if stop(pop) == True:
        print()
        print('Populasi dengan solusi optimal didapatkan')
        break
    print()
    print('Generasi ' + str(i))
    print(pop)
    


    
```
